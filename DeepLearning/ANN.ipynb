```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from keras import layers,models,Sequential 
from keras.layers import Dense
from sklearn.model_selection import cross_val_score 
```

```python
data = pd.read_csv('iris.csv')
data.head()
```

```python
data.shape
```

```python
data['class'].unique()
```

```python
X = data.iloc[:,1:4]
y =data.iloc[:,4]
```

```python
label_encoder = LabelEncoder()
y_temp = label_encoder.fit_transform(y)
Y = pd.get_dummies(y_temp).values
```

```python
Y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
from keras.optimizers import SGD,Adam
nn1 = Sequential()
nn1.add(Dense(4,input_shape=(3,),activation='relu',use_bias=True,bias_initializer='ones',kernel_initializer='random_normal'))
nn1.add(Dense(3,activation='softmax'))
nn1.compile(Adam(lr=0.01),loss="categorical_crossentropy",metrics=['accuracy'])
nn1.summary()
```

```python
model1= nn1.fit(X_train,y_train,epochs=100,batch_size=10,validation_split=0.2)
```

```python
y_pred = nn1.predict(X_test)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(model1.history['accuracy'], label='Traing_accuracy')
plt.plot(model1.history['val_accuracy'],label='Val_accuracy')
plt.xlabel('Iterations')
plt.ylabel('accuracy')
plt.title('Accuracy Level')
plt.legend()
plt.show()
```

```python
nn2 = Sequential()
nn2.add(Dense(9,input_shape=(3,),activation='relu',use_bias=True,bias_initializer='ones',kernel_initializer='random_normal'))
nn2.add(Dense(3,activation='relu'))
nn2.add(Dense(3,activation='softmax'))
nn2.compile(Adam(lr=0.01),loss="categorical_crossentropy", metrics=['accuracy'])
nn2.summary()
```

```python
model2=nn2.fit(X_train,y_train,epochs=200,batch_size=10,validation_split=0.2)
```

```python
y_pred = nn2.predict(X_test)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(model2.history['accuracy'], label='Traing_accuracy')
plt.plot(model2.history['val_accuracy'],label='Val_accuracy')
plt.xlabel('Iterations')
plt.ylabel('accuracy')
plt.title('Accuracy Level')
plt.legend()
plt.show()
```
